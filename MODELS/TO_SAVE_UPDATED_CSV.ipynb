```python
#remove Duplicates
df=df.drop_duplicates(keep='first')
```

```python
df['target'].value_counts()
```

```python
df['num_characters']=df['text'].apply(len)
df.head()
```

```python
import nltk
# Fetch Number of Words
df['num_words']=df['text'].apply(lambda x:len(nltk.word_tokenize(x)))
df.head()
```

```python
df['num_sentences']=df['text'].apply(lambda x:len(nltk.sent_tokenize(x)))
df.head()
```

```python
from nltk.corpus import stopwords
import string
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def transform_text(text):
    text=text.lower()
    text=nltk.word_tokenize(text)
    
    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
    
    text=y[:]
    y.clear()
    
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
    
    text=y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
        
    return " ".join(y)
```

```python
df['transformed_text']=df['text'].apply(transform_text)
df.head()
```

```python
#nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
df['word_tokenize']=df['transformed_text'].apply(lambda x:nltk.word_tokenize(x))
df.head()
```

```python
df['pos']=df['word_tokenize'].apply(lambda x:nltk.pos_tag(x))
df.head()
```

```python
# df.drop(columns=['Unnamed: 0','Unnamed: 0.1'],inplace=True)
df.head()
```

```python
df['verbs']=df['pos'].apply(lambda x:[word for word, tag in x if tag in ['VB','VBD','VBG','VBN','VBP','VBZ']])
df.head()
```

```python
df['adjectives']=df['pos'].apply(lambda x:[word for word, tag in x if tag in ['JJ','JJR','JJS']])
df.head()
```

```python
df['nouns']=df['pos'].apply(lambda x:[word for word, tag in x if tag in ['NN','NNS','NNP','NNPS']])
df.head()
```
